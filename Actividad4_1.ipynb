```python
#Eliminamos las primeras 5 filas porque no nos sirven
data1= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2020', skiprows=5)
data1
```

```python
data1.info()
```

```python
#Identificar valores nulos por columna
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Eliminamos las columnas 'GASTO' y 'POLIZA' ya que la mayoría de sus datos son nulos, y la Columna1 porque no especifica que es.
#También eliminamos las columnas 'RET ISR' y 'RET IVA' ya que tienen demasiados nulos pero este entorno no los detecta
data2= data1.drop(['GASTO','POLIZA', 'Columna1', 'RET ISR', 'RET IVA'], axis=1)
data2
```

```python
#Verificamos valores nulos
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Valores nulos de FOLIO
#Ya que es un numero único, colocamos "No identificado"
data2["FOLIO"] =data2["FOLIO"].fillna("NO IDENTIFICADO")
```

```python
#Verificamos valores nulos
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#NA's de TC
#Verificamos moda
TC = data2.groupby(['TC'])['TC'].count()
TC
```

```python
#Sustitucion nulos de TC
#Ya que la moda es 1 la sustituimos en los datos
data2["TC"]= data2["TC"].fillna(1)
```

```python
#Verificamos valores nulos
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion nulos de TIPO
#Ya que no sabemos a que categoria corresponde cada letra, colocamos "NO IDENTIDICADO"
#Valores nulos de FOLIO
#Ya que es un numero único, colocamos "No identificado"
data2["TIPO"] =data2["TIPO"].fillna("NO IDENTIFICADO")
```

```python
#Verificamos valores nulos
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir nulos de IMPORTE
#Ya que se trata de precios, sustituimos con la media de los importes de otras compras
data2["IMPORTE"]=data2["IMPORTE"].fillna(round(data2["IMPORTE"].mean(),1))
```

```python
#Verificamos valores nulos
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir nulos de IVA
#Ya que se trata de un valor numerico que viene del importe, sustituimos los nulos con la media
data2["IVA"]=data2["IVA"].fillna(round(data2["IVA"].mean(),1))
```

```python
#Verificamos valores nulos
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data3= data2.drop(['FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'DESCRIPCION', 'TIPO', 'STATUS' ], axis=1)
data3
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data3.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data3

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= data3[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data4_iqr= data3[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```

```python
#TC
#Primero revisamos por grupos
#ChecarTC = data4_iqr.groupby(['TC'])['TC'].count()
#ChecarTC

#Una vez que revisamos, sustituimos outliers con la moda (1)
data4_iqr["TC"]= data4_iqr["TC"].fillna(1)
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) de TOTAL SAT, TOTAL MX, IMPORTE, IVA  con "mean"
data5_iqr=data4_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data4_iqr.mean(),1))
data5_iqr
```

```python
# Unimos las columnas cualitativas con el dataframe

cualitativos=data2[['FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'DESCRIPCION', 'TIPO', 'STATUS']]
Datos_limpios = pd.concat([cualitativos, data5_iqr], axis=1)
Datos_limpios
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Datos_limpios2020.csv")
```
