```python
#Eliminamos las primeras 3 filas porque no nos sirven
data1= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2021', skiprows=3)
data1
```

```python
data1.info()
```

```python
#Identificar valores nulos por columna
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Eliminamos la columnas 'POLIZA' ya que la mayoría de sus datos son nulos
#También eliminamos las columnas 'RET ISR' y 'RET IVA' ya que tienen demasiados nulos pero este entorno no los detecta
data2= data1.drop(['POLIZA', 'RET ISR', 'RET IVA'], axis=1)
data2
```

```python
#Verificamos valores nulos
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion de nulos de FOLIO Y MP
#Ya que en folio se trata de una clave única, y MP de una ubicación (que en este caso no conocemos), colocamos en ambas variables "NO IDENTIFICADO"
data2["FOLIO"] =data2["FOLIO"].fillna("NO IDENTIFICADO")
data2["MP"] =data2["MP"].fillna("NO IDENTIFICADO")
```

```python
#Verificamos valores nulos
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data3= data2.drop(['FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'DESCRIPCION', 'TIPO', 'STATUS','MP' ], axis=1)
data3
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data3.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

MÉTODO: DESVIACIÓN ESTANDAR

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data3
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= data3[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data4= data3[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data4
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#TC
#Primero revisamos por grupos
#ChecarTC = data4.groupby(['TC'])['TC'].count()
#ChecarTC

#Una vez que revisamos, sustituimos outliers con la moda (1)
data4["TC"]= data4["TC"].fillna(1)
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Las variables IMPORTE, IVA, TOTAL MX y TOTAL SAT las sustituimos con media ya que se trata de $
data5=data4.copy()
data5=data5.fillna(round(data4.mean(),1))
data5
```

```python
# Unimos las columnas cualitativas con el dataframe

cualitativos=data2[['FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'DESCRIPCION', 'TIPO', 'STATUS', 'MP']]
Datos_limpios = pd.concat([cualitativos, data5], axis=1)
Datos_limpios
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Datos_limpios2021.csv")
```
